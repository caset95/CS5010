```python
#Wish to web scrape information about the top 100 wines of 2019:
#-rank
#-name
#-vintage(year)
#-wine score
#-price

#Make a GET request to the server to download the HTML contents of the
# Wine Spectator "Top 100 Lists"

#A response of 200 means the connection was successful
import requests 
page = requests.get("https://top100.winespectator.com/lists/")
page
```

```python
#use the BeautifulSoup library to parse this document 
#must first create an instance of the BeautifulSoup class
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
#Identify the table we want to extract data from
wineTable = soup.find('tbody')
```

```python
#Gather the rank of all the wines
ranking = [r.get_text() for r in wineTable.select(".rank")]
```

```python
#Make sure all rankings were captured
print(ranking)
```

```python
#Gather the name of the all top 100 wines
wineName = [w.get_text() for w in wineTable.select(".table-name .wineName")]
```

```python
#Make sure all wines were captured
print(wineName)
```

```python
#Use a for loop to get rid of excess characters after the name
i=0
for name in wineName:
    wineName[i] = name[:-9]
    i+=1
print(wineName)
```

```python
#Gather the year of each wine in the top 100 list
vintage = [v.get_text() for v in wineTable.select(".vintage")]
```

```python
#Make sure information was gathered
print(vintage)
```

```python
#Gather the wine score for each wine
score = [s.get_text() for s in wineTable.select(".score")]
```

```python
#Make sure the scores were gathered
print(score)
```

```python
#Gather the price of each wine
price = [p.get_text() for p in wineTable.select(".price")]
```

```python
#Make sure the prices were gathered
print(price)
```

```python
#Use a for loop to get rid of dollar sign 
#this is to make future calculations easier
i=0
for amt in price:
    price[i] = amt[1:]
    i+=1
print(price)
```

```python
#Now want to store our data in a csv file

#first must store our column data as rows
wineDataRows = zip(ranking, wineName, vintage, score, price)
    
import csv

newfilePath = "./2020WineData.csv"
#will create the open file with UTF-8 encoding to allow us to store the data in a pandas dataframe later
with open(newfilePath, "w", newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(["Rank", "Name", "Vintage", "Score", "Price(USD)"]) #add headers
    for row in wineDataRows:  #now add the web scraped data to the file
        writer.writerow(row)
```

```python
#Now want to read data from csv in order to manipulate it
#For practice, store the data in a pandas dataframe 
#can easily perform analysis with pandas library
import pandas as pd 
data = pd.read_csv('2020WineData.csv')
data
```
